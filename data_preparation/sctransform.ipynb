# Processing with Seurat dataset integration workflow

```r
seurat.obj_int <- CreateSeuratObject(counts = matrix, meta.data = col_data)
seurat.obj_int <- SplitObject(seurat.obj_int, split.by = "region")
```

```r
genes_retained<-list()
for (i in names(seurat.obj_int)) {
    seurat.obj_int[[i]] <- SCTransform(seurat.obj_int[[i]],variable.features.n=NULL,
                        return.only.var.genes = TRUE,conserve.memory=TRUE )
    genes_retained[[i]]<-rownames(GetAssayData(object = seurat.obj_int[[i]], slot = "scale.data"))
}
```

```r
features<-Reduce(intersect,genes_retained)
#features <- SelectIntegrationFeatures(object.list = seurat.obj_int, nfeatures = length(features))
```

```r
#saveRDS(seurat.obj_int,'scaled_integrated.rds')
#saveRDS(features,'scaled_integrated_selectedFeats.rds')
#seurat.obj_int<-readRDS('scaled_integrated.rds')
#features<-readRDS('scaled_integrated_selectedFeats.rds')
```

```r
seurat.obj_int <- PrepSCTIntegration(object.list = seurat.obj_int, anchor.features = features)
anchors <- FindIntegrationAnchors(object.list = seurat.obj_int, normalization.method = "SCT", 
    anchor.features =features)
seurat.obj_int <- IntegrateData(anchorset = anchors, normalization.method = "SCT")
```

```r
scale.data_int<-GetAssayData(object = seurat.obj_int, slot = "scale.data")
write.table(scale.data_int, file = "scaled_integrated_scaledata.tsv",sep='\t')
```

```r
seurat.obj_int <- RunPCA(object = seurat.obj_int,verbose=FALSE,npcs=100)
ElbowPlot(seurat.obj_int)
```

```r
seurat.obj_int <- RunUMAP(object = seurat.obj_int, dims = 1:16,verbose=FALSE,min.dist=0.1)
```

```r
saveRDS(seurat.obj_int,'scaled_integrated.rds')
#seurat.obj_int<-readRDS(paste(data_path,'scaled_integrated.rds',sep=''))
```
