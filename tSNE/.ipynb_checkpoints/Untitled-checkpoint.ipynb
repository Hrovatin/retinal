```python
d=pd.DataFrame(load_iris()['data'])

y=pd.DataFrame(load_iris()['target'])
```

```python
idx_shuffled=d.index.values.copy()
shuffle(idx_shuffled)
split1=int(len(idx_shuffled)*0.8)
split2=int(len(idx_shuffled)*0.9)
train=idx_shuffled[:split1]
validation=idx_shuffled[split1:split2]
test=idx_shuffled[split2:]
print('N train, test, validation:',len(train),len(validation),len(test))
d_t = d.loc[train,:]
d_v = d.loc[validation,:]
d_te = d.loc[test,:]
```

```python
print('\nRegularisation parameter C:',round(1,3))
```

```python
for c in arange(0.1,0.3,0.2):
    print(c)
    m=make_log_regression(d_t,d_v,col_data=y,label=0,
                        logreg={'penalty':'l1','C':c,'random_state':0,'solver':'saga','n_jobs':8})
```

```python
features=d.columns.values
for group,weights in zip(m.classes_,m.coef_):
    print(group,len(features[weights!=0]))
```
